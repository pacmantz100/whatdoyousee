```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics

normal = np.random.normal(10,5,100)
```

```python
plt.plot(normal)
plt.show()
```

```python
plt.hist(normal)
plt.axvline(normal.mean(), color = 'b', linestyle='solid', linewidth=2)

plt.axvline(normal.mean() + normal.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(normal.mean()-normal.std(), color='b', linestyle='dashed', linewidth=2) 
```
